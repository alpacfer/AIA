# Segmentation boundry

```python
def boundary_length(im):
    '''
    Computes the boundary length of an image.
    
    Parameters
    ----------
    im : ndarray
        Image.
        
    Returns
    -------
    bl : float
        Boundary length.
    '''
    return (im[:-1] != im[1:]).sum() + (im[:,:-1] != im[:,1:]).sum()
```

```python
# Load the images
dir = 'data_week1/fuel_cells/'
img1 = skimage.io.imread(dir + 'fuel_cell_1.tif')
img2 = skimage.io.imread(dir + 'fuel_cell_2.tif')
img3 = skimage.io.imread(dir + 'fuel_cell_3.tif')

# Show the images in the same figure
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(img1, cmap='gray')
ax[0].set_title('Fuel Cell 1')
ax[1].imshow(img2, cmap='gray')
ax[1].set_title('Fuel Cell 2')
ax[2].imshow(img3, cmap='gray')
ax[2].set_title('Fuel Cell 3')
plt.show()
```

```python
# Put the images in a list
imgs = [img1, img2, img3]

# Calculate the length of the segmentation boundary for each image
for img in imgs:
    print('Length of the segmentation boundary:', boundary_length(img))
```
